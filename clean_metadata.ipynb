```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from datetime import datetime

# Загружаем данные
df = pd.read_csv('metadata.csv')

# Первичный осмотр данных
print(f"Всего записей: {len(df)}")
print("\nПервые 5 строк:")
display(df.head())
print("\nИнформация о данных:")
display(df.info())
print("\nОписательная статистика:")
display(df.describe(include='all'))
print("\nКоличество пропусков по столбцам:")
display(df.isna().sum())
```

1. Обработка пропущенных значений
1.1. Заполнение пропусков в возрасте

```python
# Гистограмма текущего распределения возраста
plt.figure(figsize=(10, 6))
sns.histplot(df['age'].dropna(), bins=30, kde=True)
plt.title('Распределение возраста до заполнения пропусков')
plt.xlabel('Возраст')
plt.ylabel('Количество')
plt.show()

# Заполняем пропуски случайными значениями в диапазоне 20-80 лет
mean_age = df['age'].mean()
std_age = df['age'].std()
min_age = max(20, mean_age - 2*std_age)
max_age = min(80, mean_age + 2*std_age)

# Генерируем случайные возраста в реалистичном диапазоне
np.random.seed(42)
missing_age_count = df['age'].isna().sum()
random_ages = np.random.randint(low=int(min_age), high=int(max_age), size=missing_age_count)

df.loc[df['age'].isna(), 'age'] = random_ages

# Проверяем результат
print(f"Пропусков в возрасте после заполнения: {df['age'].isna().sum()}")

# Гистограмма после заполнения
plt.figure(figsize=(10, 6))
sns.histplot(df['age'], bins=30, kde=True)
plt.title('Распределение возраста после заполнения пропусков')
plt.xlabel('Возраст')
plt.ylabel('Количество')
plt.show()
```

1.2. Обработка пропусков в поле пола

```python
# Анализ распределения пола
gender_dist = df['sex'].value_counts(normalize=True)
print("Распределение пола:")
display(gender_dist)

# Заполняем пропуски на основе существующего распределения
missing_gender = df['sex'].isna()
df.loc[missing_gender, 'sex'] = np.random.choice(
    gender_dist.index, 
    size=missing_gender.sum(), 
    p=gender_dist.values
)

print(f"Пропусков в поле пола после заполнения: {df['sex'].isna().sum()}")
```

2. Унификация диагнозов

```python
# Анализ уникальных диагнозов
print("Уникальные диагнозы до обработки:")
display(df['finding'].value_counts())

# Функция для унификации диагнозов
def unify_diagnosis(diagnosis):
    diagnosis = str(diagnosis).lower().strip()
    if 'covid' in diagnosis or 'ards' in diagnosis:
        return 'COVID-19'
    elif 'pneumonia' in diagnosis:
        return 'Pneumonia'
    elif 'sars' in diagnosis:
        return 'SARS'
    elif 'normal' in diagnosis or 'no finding' in diagnosis:
        return 'Normal'
    else:
        return diagnosis.capitalize()

# Применяем унификацию
df['finding'] = df['finding'].apply(unify_diagnosis)

# Проверяем результат
print("\nУникальные диагнозы после обработки:")
display(df['finding'].value_counts())
```

3. Удаление дубликатов

```python
# Проверка на полные дубликаты
print(f"Всего записей до удаления дубликатов: {len(df)}")
duplicates = df.duplicated()
print(f"Найдено полных дубликатов: {duplicates.sum()}")

# Удаляем дубликаты
df = df.drop_duplicates()
print(f"Всего записей после удаления дубликатов: {len(df)}")

# Проверка на дубликаты по ID пациента (если есть)
if 'patientid' in df.columns:
    patient_duplicates = df['patientid'].duplicated()
    print(f"Найдено дубликатов patientid: {patient_duplicates.sum()} (есть patientid, у которых по несколько снимков, оставляем)")
   
```

4. Обработка аномальных значений
4.1. Анализ возраста

```python
# Выявляем аномальные значения возраста
Q1 = df['age'].quantile(0.25)
Q3 = df['age'].quantile(0.75)
IQR = Q3 - Q1
lower_bound = Q1 - 1.5 * IQR
upper_bound = Q3 + 1.5 * IQR

outliers = df[(df['age'] < lower_bound) | (df['age'] > upper_bound)]
print(f"Найдено возрастных выбросов: {len(outliers)}")

# Визуализация
plt.figure(figsize=(10, 6))
sns.boxplot(x=df['age'])
plt.title('Распределение возраста с выбросами')
plt.show()

# Обработка выбросов
df['age'] = df['age'].clip(lower=lower_bound, upper=upper_bound)

# Визуализация после обработки
plt.figure(figsize=(10, 6))
sns.boxplot(x=df['age'])
plt.title('Распределение возраста после обработки выбросов')
plt.show()
```

4.2. Анализ других полей

```python
# Анализ поля 'view' (проекция снимка)
print("\nУникальные значения проекции снимка:")
display(df['view'].value_counts())

# Можно унифицировать значения
df['view'] = df['view'].str.upper().str.strip()

if 'date' in df.columns:
    try:
        # Сохраняем оригинальные даты для проверки
        df['date_original'] = df['date']
        
        # Пробуем преобразовать в datetime
        df['date'] = pd.to_datetime(df['date'], errors='coerce')
        
        # Анализируем некорректные даты
        invalid_dates = df['date'].isna()
        print(f"\nНекорректных дат: {invalid_dates.sum()}")
        
        if invalid_dates.any():
            # Находим моду (наиболее частую дату)
            mode_date = df['date'].mode()[0]
            print(f"\nНаиболее частая дата в данных: {mode_date}")
            
            # Заполняем некорректные даты модой
            df.loc[invalid_dates, 'date'] = mode_date
            
            # Проверяем результат
            print(f"После заполнения некорректных дат:")
            print(f"Осталось некорректных дат: {df['date'].isna().sum()}")
            
            # Дополнительный анализ
            print("\nПримеры исправленных записей:")
            display(df.loc[invalid_dates, ['date_original', 'date']].head())
    except Exception as e:
        print(f"\nОшибка при обработке дат: {e}")
        # В случае ошибки оставляем даты как есть
        if 'date_original' in df.columns:
            df['date'] = df['date_original']
```

5. Создание новых аналитических признаков
5.1. Возрастные группы

```python
# Определяем возрастные группы
bins = [0, 18, 65, 120]
labels = ['Дети', 'Взрослые', 'Пожилые']
df['age_group'] = pd.cut(df['age'], bins=bins, labels=labels, right=False)

# Проверяем распределение
print("\nРаспределение по возрастным группам:")
display(df['age_group'].value_counts())
```

8. Распределения изображений по папкам в соответствии с диагнозами

```python
import os
import shutil
import pandas as pd

# Загрузка метаданных
metadata = pd.read_csv('metadata_cleaned_.csv')

# Проверка наличия необходимых столбцов
if 'filename' not in metadata.columns or 'finding' not in metadata.columns:
    raise ValueError("Файл metadata_cleaned_.csv должен содержать столбцы 'filename' и 'finding'")

# Пути к папкам
source_dir = 'images'  # Папка с исходными изображениями
dest_base_dir = 'sorted_images'  # Базовая папка для сортировки

# Создаем базовую папку, если ее нет
os.makedirs(dest_base_dir, exist_ok=True)

# Счетчики
total_files = 0
moved_files = 0
missing_files = 0

# Проходим по всем записям в metadata
for index, row in metadata.iterrows():
    filename = row['filename']
    diagnosis = row['finding']
    
    # Пропускаем записи без имени файла или диагноза
    if pd.isna(filename) or pd.isna(diagnosis):
        continue
    
    # Создаем безопасное имя папки (удаляем спецсимволы)
    safe_diagnosis = "".join(c if c.isalnum() else "_" for c in str(diagnosis))
    diagnosis_dir = os.path.join(dest_base_dir, safe_diagnosis)
    
    # Создаем папку для диагноза, если ее нет
    os.makedirs(diagnosis_dir, exist_ok=True)
    
    # Полные пути к файлам
    src_path = os.path.join(source_dir, filename)
    dest_path = os.path.join(diagnosis_dir, filename)
    
    total_files += 1
    
    # Проверяем существование исходного файла
    if os.path.exists(src_path):
        # Копируем файл (используйте shutil.move для перемещения вместо копирования)
        shutil.copy2(src_path, dest_path)
        moved_files += 1
    else:
        missing_files += 1
        print(f"Файл не найден: {filename}")

# Отчет
print(f"\nОбработка завершена:")
print(f"Всего записей: {len(metadata)}")
print(f"Обработано файлов: {total_files}")
print(f"Успешно перемещено: {moved_files}")
print(f"Отсутствующих файлов: {missing_files}")
print(f"\nФайлы распределены по папкам в директории: {dest_base_dir}")
```

9. Удаления строк из metadata_cleaned_.csv, которые ссылаются на отсутствующие изображения

```python
import os
import pandas as pd

# Пути к файлам и папкам
metadata_path = 'metadata_cleaned_.csv'
images_dir = 'images'  # Папка с изображениями
output_csv_path = 'metadata_cleaned.csv'  # Выходной файл CSV
output_parquet_path = 'metadata.parquet'  # Выходной файл Parquet

# Загрузка метаданных
metadata = pd.read_csv(metadata_path)

# Проверка наличия столбца с именами файлов
if 'filename' not in metadata.columns:
    raise ValueError("Файл metadata_cleaned_ должен содержать столбец 'filename'")

# Функция для проверки существования файла
def check_file_exists(filename):
    if pd.isna(filename):
        return False
    return os.path.exists(os.path.join(images_dir, filename))

# Создаем маску для существующих файлов
valid_files_mask = metadata['filename'].apply(check_file_exists)

# Разделяем данные
existing_files_metadata = metadata[valid_files_mask]
missing_files_metadata = metadata[~valid_files_mask]

# Сохраняем статистику
total_entries = len(metadata)
existing_entries = len(existing_files_metadata)
missing_entries = len(missing_files_metadata)

# Сохраняем очищенные метаданные в CSV
existing_files_metadata.to_csv(output_csv_path, index=False)

# Сохраняем очищенные метаданные в Parquet
existing_files_metadata.to_parquet(output_parquet_path, index=False)

# Сохраняем список отсутствующих файлов (для отчета)
missing_files_metadata.to_csv('missing_files_report.csv', index=False)
missing_files_metadata.to_parquet('missing_files_report.parquet', index=False)

# Выводим отчет
print(f"Обработка завершена:")
print(f"Всего записей в исходном файле: {total_entries}")
print(f"Найдено существующих файлов: {existing_entries}")
print(f"Обнаружено отсутствующих файлов: {missing_entries}")
print(f"\nОчищенные метаданные сохранены в:")
print(f"CSV: {output_csv_path}")
print(f"Parquet: {output_parquet_path}")
print(f"\nОтчет по отсутствующим файлам сохранен в:")
print(f"CSV: missing_files_report.csv")
print(f"Parquet: missing_files_report.parquet")

# Дополнительная проверка: выводим первые 5 отсутствующих файлов
if missing_entries > 0:
    print("\nПримеры отсутствующих файлов:")
    print(missing_files_metadata['filename'].head().to_string(index=False))
```
